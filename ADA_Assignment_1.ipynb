##Innovation 

The concept of the generative model belongs to probability statistics and machine learning and refers to a series of models used to randomly generate observable prediction data. In short, the "generated" sample is as similar as possible to the "true" sample. The two main functions of the generative model are to learn a probability distribution and to generate data, which is very important and can be used not only in unsupervised learning but also in supervised learning.
 
This paper introduced a new procedure which is an adversarial model. This procedure is the biggest innovation of the generation adversarial network compared with previous generation models such as an automatic encoder. The adversarial procedure is simply a discriminative model for true or false, which is a dichotomy problem, where you put in a true image and you expect the discriminative model to give you a 1, and you put in a false image and you expect the discriminative model to give you a 0. This has nothing to do with the label of the original picture. No matter how many categories the original picture is, they are all called as true pictures. If the output label is 1, it means that it is true. The label of the generated image is 0, which means false.
 
In the training process, the discriminative model is trained first, and the false data and the true data are input into the discriminant model. At this time, the discriminative model is optimized, hoping that it can correctly judge the true data and the false data, so that a better discriminative model can be obtained.
 
Then we start training the generative model, hoping that the fake data it generates will fool the better discriminative model it has now. The particular way is to distinguish parameters are fixed, the parameters of the generative model were optimized by backpropagation, we expect the generative model with the data after judging device results can be as much as possible close to 1. Then we just need to adjust the loss function, the optimal discriminant apparatus before the loss function is to make the false data as close as possible to zero, and now the trained the generative model loss function is to make the false data as close to 1, so that we could let the generative model generate the expected result. (Hui, J. 2019)

And also, this paper uses mathematical expression to explain the idea of adversarial network: 

![mathematical Function](https://github.com/RobertMa05/ADA-Assignment-1/blob/master/%E5%B1%8F%E5%B9%95%E6%88%AA%E5%9B%BE(163).png?raw=true)

$V(G,D)$ represents the degree of difference between the real sample and the generated sample.

$max_{D}[V(G,D)]$ means fixed generator G. the discriminator can maximize whether the sample comes from real data or generated data as much as possible.

$min_{G}⁡max_{D}[V(G,D)]$ is to obtain generator G under the condition of fixed discriminator D, which requires to minimize the difference between the real sample and the generated sample.



With this minimax game, it will converge to the generated distribution and fit to the real distribution. (Hong et al. 2018)

With this set of distribution diagrams and the mathematical function, it is easier to understand how the generator and the discriminator work together step by step, so that we can get the expected result from the entire generation adversarial model.
